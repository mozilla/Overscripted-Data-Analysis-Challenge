
## Browser Attribute Fingerprinting Scripts Analysis 

There are some scripts that we can pick out by name that are doing browser attribute fingerprinting:
    
1. Scripts with hs-analytics in the script_url
2. Scripts with /akam/ in the script_url
3. The fingerprintjs scripts

This notebook has used the existing method which @birdsarah has produced in analyses/issue_34_setup_and_dask_tips.ipynb to pull out these scripts.

It calculates the percentage of each of the three scripts with respect to the total number of scripts. It also calculates the percentage when the three of the scripts are considered together.

Value_1000_only dataset is used which contains all the rows of the dataset, but truncates the value field to only keep the first 1000 characters in a column called value_1000. 


```python
import dask.dataframe as dd
from dask.distributed import Client

Client()
```

```python
df = dd.read_parquet(
    '/home/soumya/Dataset/sample_10percent_value_1000_only.parquet', engine='pyarrow',
    columns=['argument_0', 'symbol', 'location', 'script_url', 'value_1000']
)
part_0 = df.get_partition(0)
print(f'''
    Complete dataset is {len(df):,} rows.
    Partition is {len(part_0):,} rows.
    Type of dataset is {type(df)}.
    Type of part_0 is {type(part_0)}.
''')
df.compute()
```

```python
# Get hs-analytics scripts
hs = df[df.script_url.str.contains('hs-analytics')]
hs.compute()
```

```python
total=11292867
hs_analytics=13848
Percentage= ((hs_analytics/total)*100)
Percentage
```

```python
# Get akam scripts
akam = df[df.script_url.str.contains('/akam/')]
akam.compute()
```

```python
total=11292867
akamscripts=60975 
Percentage= ((akamscripts/total)*100)
Percentage
```

```python
# Get fingerprintjs2 scripts
fp2 = df[df.script_url.str.contains('fingerprint2')]
fp2.compute()
```

```python
total=11292867
fingerprintjs2=10649
Percentage= ((fingerprintjs2/total)*100)
Percentage
```

Let's finish by looking for all the different ways fingerprintjs2 (or scripts very like it) may be referenced in the dataset.  Instead of just looking for fingerprintjs2 in the name we can look for the characteristic value `Cwm fjordbank glyphs vext quiz`

```python
fp_arguments = df[df.argument_0.str.contains('Cwm fjordbank glyphs vext quiz')]
fp_arguments.compute()
```

```python
tal=11292867
fp_script=10649
Percentage= ((fp_script/total)*100)
Percentage
```

## Conclusion

It can be seen, out of 11292867 scripts  13848 contain hs_analytics, 60975 akam, 10649 fingerprintjs in the url which is 0.123%, 0.539%, 0.094% of the total scripts respectively.

Amongst the three akam scripts have the greatest contribution of 0.539%.

If we consider three of the scripts together it forms 0.756% of the total scripts.
